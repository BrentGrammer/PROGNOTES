```python
### PERLIN NOISE ###

import numpy as np
import matplotlib.pyplot as plt

# Controlled randomness (smooth out random steps to generate more terrain like randomness)
  # see https://eev.ee/blog/2016/05/29/perlin-noise/


## Linear Interpolation ##
#  if you’re t of the way between two extremes A and B, then the linear interpolation is A + t(B - A) or, equivalently, A(1 - t) + Bt. For t = 0, you get A; for t = 1, you get B; for t = ½, you get ½(A + B), the average.

# Random slopes to start with:
x = np.arange(0, 10)
rng = np.random.default_rng() # more modern way of making the numbers random with statistical advantages from just np.random.uniform
slopes = rng.uniform(-1, 1, len(x)) # based on np.random.uniform - uniformly distributed random numbers (all equal probability)
print(slopes)

print('not with rngdefault:')
print(np.random.uniform(-1,1,len(x))) # uses the older Mersenne Twister algorithm, which is still widely used but has some known statistical weaknesses.
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot points
ax.scatter(x, np.zeros_like(x), color='black', zorder=3)

# Plot slope lines
for i, slope in enumerate(slopes):
    ax.plot([i, i+1], [0, slope], color='lightblue', zorder=2)

# Set plot limits and labels
ax.set_xlim(-0.5, len(x) - 0.5)
ax.set_ylim(-1.1, 1.1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Random Slopes at Integer Points')

# Add grid
ax.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
