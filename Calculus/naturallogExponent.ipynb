```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

n = [1,2,5,10]

for i in n:
    # formula for estimating the natural exponent e
    e = (1+(1/i))**i
    # formating with 2.0f means 2 significant digits to the left of the decimal and 0 digits to the right of the decimal point (just for aligning)
    print(f'n: {i:2.0f}, estimated e: {e: .5f}, diff to actual e: {np.exp(1)-e:.5f}') # e to power of 1 is just e - we compare the estimated calculations to the numpy estimate of natural exponent

# note how the printed statements show that the interation using the formula to estimate e is approaching the numpy estimation of e
```

```python
# visualize the iteration as it approaches e

# use a vector of n's to repeat 1000 times
n = np.arange(1,1001) # exclusive end, so 1001 makes 1000 total

eDiffs = np.exp(1) - (1 + 1/n)**n # this runs on the arange above element wise (each element)

plt.plot(n,eDiffs)

# note how after (less than) 100 iterations the value converges quickly to the more accurate numpy e
  # the iterations from 100 to 1000 asymptotically approach 0.
```

```python
xDomain = [-2,2]

# get set of numbers in the range of the domain above
x = np.linspace(xDomain[0], xDomain[1], 53) # arbitrary number of steps in set

# define various functions of the natural exponent to plot
y1 = np.exp(x) # e^x
y2 = np.exp(x**2) #e^(x^2)
y3 = np.exp((-x)**2) # e^(-x^2)
y4 = np.exp(-(x**2)) # e^-x^2
y5 = np.exp(x)**2 # (e^x)^2

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.plot(x,y5)

plt.ylim([-1,10]) # reduce size of y axis
```

```python
# sympy vs. numpy `e`/using sympy and numpy in the same expression

s_beta = sym.var('beta')

# function using sympy and numpy in same expression
  # we subtract log to eliminate a dip below 0 on the beta (x) axis in the plot
s_y = sym.exp(s_beta) - sym.log(s_beta) -  np.exp(1) # sympy knows that np.exp is not actually e fully, and just treats it as a number

xDomain = [-2,2]

# sympy will fill in beta with numbers (i.e. 0 to 10) and plot
sym.plot(s_y, (s_beta, xDomain[0], xDomain[1]),
         title=f'$f(\\beta) = {sym.latex(s_y)}$') # set the domain for the x-axis/beta
# Note that our domain for the function is all positive and we are passing in a -2 as the start of x domain (sympy just ignores invalid part of the domain)
```

```python
# Demonstrate that the natural log and natural exponent are inverses of each other

x = np.linspace(-4,4,30)

# e^x for the range -4 to 4
plt.plot(x, np.exp(x), linewidth=2, label='$e^x$')


# log(e^x)
plt.plot(x, np.log(np.exp(x)), linewidth=2, label='$log(e^x)$')

# inverse of above: e^(log(x))
  # note we will get a warning because we are computing the natural log of negative numbers
plt.plot(x, np.exp(np.log(x)), 'o', linewidth=2, label='$e^{log(x)}$')


# we need a new vector for x to prevent skipping numbers close to 0 and not getting the full range of y
x = np.linspace(.001,4,30)
# log of x
plt.plot(x, np.log(x), linewidth=2, label='$log(x)$')

plt.legend()
plt.axis('square') # makes plot symmetrical in display (squares it up)
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.xlabel('x')
plt.ylabel('y')
plt.show()


# Note how the log(e^x) and e^log(x) cancel each other out and the line is just a straight diagonal line (x = y)
```
