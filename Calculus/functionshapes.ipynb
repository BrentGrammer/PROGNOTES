```python
#####  SKETCHING FUNCTIONS (FUNCTION SHAPES INTUITION) ####

import numpy as np
import sympy as sym
import matplotlib.pyplot as plt


# Generate a bunch of functions to practice sketching

# 2nd order polynomials to sketch:

# generate 3 coefficients between -5 and +5 (remember randint has exclusive upper bound, so we pass in 6)
coefs = np.random.randint(-5,6,3) # generate static set of 3 coefficients between -5 and 5, will be different each time it's run

# define 101 x values we can use between -3 and 3
x = np.linspace(-3,3,101)

# create 2nd order polynomial - fx holds all the results of this equation given all the x inputs above
fx = coefs[0] + coefs[1]*x + coefs[2]*x**2 # a + bx + cx^2 - [...results for every x]

# print the function to practice sketching
print(f'Sketch this:\nf(x) = {coefs[0]} + {coefs[1]}x + {coefs[2]}x**2')

# plot the function to check your estimation of the shape
plt.plot(x, fx, label='f(x)') # plot x by f(x)

# plot some easy x-axis points to plot (as you'd do when generating a sketch)
xpnts = [-1,0,1]
markers = ['ro', 'gs', 'mp'] # red circle, green square, magenta pentagram
# zip markers and points so you can access the corresponding marker for the point in a pair tuple
for xi, mi in zip(xpnts, markers):
    y = coefs[0] + coefs[1]*xi + coefs[2]*xi**2
    plt.plot(xi, y, mi, label=f'When x={xi}')

plt.legend()

plt.grid() 
plt.show()

# note the behavior based on the coefficients generated.  
  # This type of function generates a parabola shape
  # If negative coefficient was used, then the parabola will open downwards
```

```python
# Generate Functions using natural exponents and cosine to sketch
from IPython.display import Math

# Use Sympy

t = sym.var('theta') # sets the variable t to the symbolic theta

# generate coefficients between -3 and 3, but exclude 0 (too easy to sketch) 
#  use numpy's choice to randomly choose a number from a list passed in so we can generate nums and check if 0, then skip
coef = np.random.choice([i for i in range(-3,4) if i != 0])

sfx = coef*sym.exp(-t) * sym.cos(2*sym.pi*t)

print('Sketch this function:')
# Can use Math or display to print out latek code
Math('f(\\theta) = ' + sym.latex(sfx)) # \theta is latek for the symbol, convert sfx to latex text using Sympy 

# Lambdify the sympy function so that we can move to plotting with matplotlib instead of sympy's built in plot
y = sym.lambdify(t,sfx)

# plot with matplotlib instead of sympy
plt.plot(x, y(x)) # plot and reuse x input against the lambdified sympy function

plt.plot(0, y(0), 'ro') # plots function when x=0 with a red dot

plt.grid(color=[.8,.8,.8],linestyle='--') # can style the grid to lighter grey and change line style to be dashed lines
plt.show()
```

```python
# Sketching a randomly generated piecewise function

x = np.linspace(-3,3,101)

negativeX = ['x','x**2','x**3'] # which rule applies when domain is negative
positiveX = ['np.log','np.sqrt'] # rules that can apply when domain is positive

# randomly select an element from each list for the piecewise function
randFunIdx = [ np.random.choice(np.arange(len(negativeX))), # pass list of 0 to length of each domain to choice
               np.random.choice(np.arange(len(positiveX))) ]
fx = eval( negativeX[randFunIdx[0]] ) # evaluating the negativeX selected rule against x values defined further above

# fx[x>0] means take the results that are only positive - (not sure why)
fx[x>0] = eval(positiveX[randFunIdx[1]] + '(x[x>0])')

# run for only positive x values of the function:
fx[x>0] = eval(positiveX[randFunIdx[1]] + '(x[x>0])') # concatenate passing in x to the rand selected function text from positiveX list. Select only values of x that are positive and more than 0 

print('Sketch this:\n')
print(f'{negativeX[randFunIdx[0]]} for x<=0')
print(f'{positiveX[randFunIdx[1]][3:]}(x) for x>0') # cutoff np. chars in np.log for print with [3:] - print from idx 3 to the end

# plot the piecewise function generated

# plot each piece of the function separately
plt.plot(x[x<=0], fx[x<=0],label=negativeX[randFunIdx[0]])
plt.plot(x[x>0], fx[x>0],label=f'{positiveX[randFunIdx[1]][3:]}(x)')

plt.legend()
plt.show()
```
