```python
import numpy as np
import matplotlib.pyplot as plt

### Create a vector (2 ways)

## Create a list - easy for simple cases, basic plotting, etc.
v2 = [ 3,-2 ] # 2-dimensional vec
v3 = [ 4,-3,2 ] # 3 dimensional vector

## Create a numpy array for more advanced operations

### Transpose a Vector from Row to a Column vector, or vice versa:
v3t = np.transpose(v3) # swaps rows and columns
# If a numpy array, just write: `v3t.T`


### Plot a vector using a line:
plt.plot([0,v2[0]],[0,v2[1]]) # plot v2 in standard position (origin 0,0)

# some extra stuff to make the plot look nicer:
plt.axis('equal')
plt.plot([-4,4],[0,0],'k--') # dotted lines along x and y
plt.plot([0,0],[-4,4],'k--')
plt.grid()
plt.axis((-4,4,-4,4)) # not sure what this does - sizing?
plt.show()
```

```python
#### Plotting a 3D Vector #####

fig = plt.figure(figsize=plt.figaspect(1)) # not sure what this does
ax = plt.axes(projection = '3d')
# plot the x,y,z lines (from origin in standard position)
ax.plot([0, v3[0]], [0, v3[1]], [0, v3[2]], linewidth=3)

# some extras for nicer looking plot - add dotted lines to help to see the line against 3d space
ax.plot([0, 0],[0, 0],[-4, 4],'k--')
ax.plot([0, 0],[-4, 4],[0, 0],'k--')
ax.plot([-4, 4],[0, 0],[0, 0],'k--')

plt.show()
```

```python
############ ADDING AND SUBTRACTING VECTORS ###############

# use numpy arrays so we can use element-wise addition (not concatenation as with lists and `+`)
v1 = np.array([ 3,-1 ])
v2 = np.array([ 2,4 ])

# add with element-wise addition with numpy:
v3 = v1 + v2

# plotting the vector addition
plt.plot([0, v1[0]],[0, v1[1]], 'b', label='v1')
# offset v2 with v1 values on the plot - note that v1[0] adds that val to both elements in v2 etc.
plt.plot([0, v2[0]] + v1[0], [0, v2[1]] + v1[1], 'r', label='v2') # shown as red line (added onto v1)
plt.plot([0, v3[0]],[0, v3[1]], 'k', label='v1+v2') # vector sum shown as black line (the TAIL of v1 to the HEAD of v2)

plt.legend()
plt.axis('square')
plt.axis((-6,6,-6,6))
plt.grid()
plt.show()
```

```python
################# SCALAR MULTIPLICATION OF VECTORS #######################

# 2d vector and scalar
v1 = np.array([ 3,-1 ])
l = -.3 # scalar 'l' for lambda, vector will become shorter (.3 is b/w 0 and 1) and flip 180 deg in "direction"
v1m = v1*l # multiplied by lambda scalar

# plot the scaled vector
plt.plot([0, v1[0]], [0, v1[1]], 'b', label='v_1')
plt.plot([0, v1m[0]], [0, v1m[1]], 'r:', label='$\lambda v_1$') # red dotted line shows result

# dynamically set x and y axis limits - set it to the higher of vector values (multiplied or not)
axlim = max([abs(max(v1)), abs(max(v1m))]) * 1.5 # scale up a bit with 1.5
plt.axis((-axlim,axlim,-axlim,axlim))


plt.grid()
plt.show()
```

```python
############ HADAMARD ELEMENT-WISE MULTIPLICATION OF VECTORS ##########

w1 = [1,3,5]
w2 = [1,3,5]

# use np.multiply function to implement element wise multiplication between the inputs
  # the inputs can be lists, vectors or matrices etc.
w3 = np.multiply(w1,w2)
print(w3)
```

```python
############### THE UNIT VECTOR ##################

# Vector with a Length of 1
## Computing the Unit Vector transformation of another Vector:

v1 = np.array([-3,6]) # 2-D to let us visualize on 2D plane

# the scalar to get the unit vector (length=1) by taking the reciprocol
mu = 1/np.linalg.norm(v1) # reciprocol of the length(norm) of the vector - 1 / ||v||

# mu is going to be smaller than 1 because we need to scale back the length of the vector to get to 1 (the v1 vec is length longer than 1)

# scale a new vector, scaled by mu computed above
v1n = v1*mu # v1 norm (length)

plt.plot([0, v1n[0]], [0, v1n[1]], 'r', label='v1-norm (Unit Vector)', linewidth=5) # make thick linewidth since it overlaps blue line to see it
plt.plot([0, v1[0]], [0, v1[1]], 'b', label='v1 (orig. vec.)')

plt.axis('square')
plt.axis(( -6,6,-6,6 ))
plt.grid()
plt.legend()
plt.show()


#### Getting the Unit Vector against a Vector with Length < 1 #####

v1 = np.array([-0.3,0.6]) # length of this is less than 1

# the scalar to get the unit vector (length=1) by taking the reciprocol
mu = 1/np.linalg.norm(v1) # reciprocol of the length(norm) of the vector - 1 / ||v||

# mu is going to be GREATER than 1 to convert v1 (less than 1) to a Unit Vector with len 1

v1n = v1*mu

plt.plot([0, v1n[0]], [0, v1n[1]], 'r', label='v1n(Unit Vector Stretched)', linewidth=5) # make thick linewidth since it overlaps blue line to see it
plt.plot([0, v1[0]], [0, v1[1]], 'b', label='v1(len < 1)')

plt.axis('square')
plt.axis(( -6,6,-6,6 ))
plt.grid()
plt.legend()
plt.show()
```

```python
############### DOT PRODUCT OF UNIT VECTORS ##############

# 2 random vectors:
n = 4 # four dimensions
v1 = np.round( 20*np.random.randn(4) ) # multiply by arbitrary 20 to spread the numbers out more (if they were just 0-4, wouldn't be much variety)
v2 = np.round( 20*np.random.randn(4) )

# Lengths of the individual vectors - square root of the dot prod. of the vector with itself |v| = sqrt{v^T v} - remember similarity to pythagorean theorem
v1m = np.sqrt(np.dot(v1,v1))
v2m = np.sqrt(np.dot(v2,v2))
# magnitude of the dot product of the vectors to compare with the lengths of the two orig. vectors:
dpm = np.abs(np.dot(v1,v2)) # ignore the size with absolute val - we just want the size of the dot product whether negative or positive

print(f'len of v1: {v1m}')
print(f'len of v2: {v2m}')
print(f'magnitude of their dot product: {dpm}') # note how much larger the mag of the dp is compared to the length of these vecs. alone
print('\n')

# normalize vectors - create new unit vectors for the original vectors in the same dir., but have length of 1
  # NOTE: dividing v1 by v1m is the same as multiplying v1 by (1/v1m)
    # (dividing by n is the same thing as multiplying by 1/n, which is also the same thing as multiplying by n^{-1})
v1u = v1/v1m # equates to v1u=(1/v1m)*v1 which follows the formula for what to scale a num by to get 1
v2u = v2/v2m
# the length of both of these vectors is now 1
print('Now the lengths of the normalized unit vectors should be 1 (or close to it due to rounding)')
print(np.sqrt(np.dot(v1u,v1u)))
print(np.sqrt(np.dot(v2u,v2u)))

print('\n')

print('Normalized Unit Vectors:')
print(v1u,v2u)
print('\n')

# the magnitude of the dot product of the unit vectors:
dpm = np.abs(np.dot(v1u,v2u))
print(f'magnitude of dot product for unit vectors: {dpm}') # Note how the mag. of this dot prod. for the unit vectors is smaller than 1


### RELATIONSHIP OF DOT PRODUCTS FOR VECTORS AND UNIT VECTORS #####

# The dot product for the original vectors was larger, while the dot prod. magnitude for their corresponding unit vectors was smaller (less than 1)
 
```

```python
############### SPANS ###################
import numpy as np
import matplotlib.pyplot as plt

# Define a set of vectors that make up a span (set S)
S1 = np.array([1,1,0]) # a vector
S2 = np.array([1,7,0]) # a vector

# Vectors that we want to find whether they are in the span defined above:
V = np.array([1,2,0])
W = np.array([3,2,1])

### Geometric Interpretation of V or W being or not being in the span of vectors S1 and S2 (set S)

# Plot all 4 of the vectors (the span set S vectors and V and W)
fig = plt.figure()
ax = plt.axes(projection = '3d') # allow for 3d plotting
ax.plot([0, S1[0]], [0, S1[1]], [.1, S1[2]+.1], 'r', linewidth=3) # x,y,z for first vector in span set S - the .1 offset for the z coord elevates the line to make it a little more visible on the plane
ax.plot([0, S2[0]], [0, S2[1]], [.1, S2[2]+.1], 'r', linewidth=3) # x,y,z for second vector in span set S - the .1 offset for the z coord elevates the line to make it a little more visible on the plane

# plot the vectors V and W
ax.plot([0, V[0]], [0, V[1]], [.1, V[2]+.1], 'g', linewidth=3) # vector V green line is on the same plane as the span set S
ax.plot([0, W[0]], [0, W[1]], [.1, W[2]+.1], 'b') # blue line for vector W not on the span plane

# Draw a plane to make it visible which vectors are on the same plane or outside of it (W)
xx, yy = np.meshgrid(range(-15,16), range(-15,16))
cp = np.cross(S1,S2) # use the cross product to define the plane of 2 independent vectors (in set S here)
print(f'cross product of span set S: {cp}')

####
# comes from the formula for defining a plane based on a vector that is perpendicular to the plane.
# You can define a plane as aX + bY + cZ = 0. Then solve for Z.
# The original equation is
# ax+by+cz = 0
# Then the code solves for z.
z1 = (-cp[0]*xx - cp[1]*yy)*1./cp[2] # ?? somehow z coordinate on the plane for the meshgrid is 0, but not sure how that is

print(1/cp[2]) # ?
ax.plot_surface(xx,yy,z1, color='lightgray') # the plane is shown for the span set S to compare where vectors V and W fall on it

# Note that vector V (green line) is on the same plane plotted as set span S (red lines), while vector W (blue line) is not on the same plane and going off in a different direction

plt.show()
```
