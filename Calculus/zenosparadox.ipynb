```python
import numpy as np
import matplotlib.pyplot as plt

# a function to use in the exercise: cos^2(x^2) + pi
def fx(u):
    return np.cos(u**2)**2 + np.pi
# x axis grid
xx = np.linspace(-2.1,2.1,211) # range with number of points in the range (211)

# xx points on x axis and results of the function for x point on y axis
plt.plot(xx, fx(xx)) # note this does not show iteration, just what value each x point results in if plugged into the function (y axis)
plt.grid()
plt.show()
```

```python
# Algorithm inspired by Zenos paradox to understand concept of the limit

# target x axis location
a = 1

# starting x axis value (x0)
x0 = np.array([a-1,a+1]) # pass in starting vals to the function from the left and from the right - we approach target a from each side (above and below)

iterations = 10
limitvals = np.zeros((iterations, 2)) # 10x2 matrix of the result of each function call (the cols are the two values, below target a and above a)
xAxisVals = np.zeros((iterations,2)) # average of two vals in function for each call of the function (it's x axis point)

# zeno's paradox algorithm

for i in range(iterations):
    # build a 10x2 matrix (limitvals) - each row is an iteration
    xAxisVals[i,:] = x0 # assigns the average of left/right to the 2 cols
    # compute and store x0 and y=f(x0) - i is row, : is the two cols in the matrix
    limitvals[i,:] = fx(x0) # store the two values of the function in row i
    
    # update the x-values - the avg of x0 and a by the two vals (left and right)
    x0 = (x0+a)/2 # x0 + a effectively becomes [0+1, 2+1] or [1, 3], then /2 operates on each: [(0 + 1) / 2, (2 + 1) / 2]


# print results in table

# [:,0] means all rows and the first column
stacked = np.vstack((xAxisVals[:,0], limitvals[:,0])) # vertically stacks x axis values first, then followed by y axis values
# now transpose the stacked values to get the x vals on the left and the right vals adjacent to them on the right
transposedLeft = stacked.T
print('\nLimit from the left is:')
print(transposedLeft)

stacked = np.vstack((xAxisVals[:,1], limitvals[:,1])) # now get the x and y values from the right (second col)
# now transpose the stacked values to get the x vals on the left and the right vals adjacent to them on the right
transposedRight = stacked.T
print('\nLimit from the right is:')
print(transposedRight)

print(f'\nValue of f(x) at x={a}:')
print(fx(a))
```

```python
#plot function as black line
plt.plot(xx,fx(xx), 'k')
plt.plot([a,a],[np.pi,1+np.pi], 'k--') # [xvals], [yvals]
plt.plot(xAxisVals,limitvals,'o',markerfacecolor='w') # python makes different color borders for each column set (xAxisVals and limitvals are each sets of 2 columns)

plt.title(f'f({a}) = {fx(a)}')
# can zoom in using xlim
plt.xlim([a-1,a+1]) # less numbers adding/subtracting zooms in further
plt.show()

# this shows the points of each iteration getting closer and closer to the target a
# you can increate the iterations number higher and higher and the points will never actually reach target a, but keep getting infinitessimally closer
# This demonstrates the idea of limits
```

```python
# basic iteration
def fx(u):
    return np.cos(u**2)**2 + np.pi

iterations = 7

seed = 24

x = np.linspace(1, iterations, iterations, dtype=int)
y = np.zeros(iterations)
input = seed

for i in range(iterations):
    output = fx(input)
    y[i] = output
    input = output


plt.grid()
plt.plot(x,y)
```
