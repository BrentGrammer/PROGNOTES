```python
# Laws/Properties of Exponents
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

# For Latek printing import these
from IPython.display import display,Math
```

```python
# import terms/symbols to use in calculations (i.e. the variables)
from sympy.abc import x,a,b 

# In this example we can use sympy to find simplifications of equations if needed (and show this property of exponents)

# demonstrating x^a * x^b = x^(a+b)
xa = x**a
xb = x**b 

# the product:
xab = xa * xb
# use imported display from IPython to print nice LaTek
display(xab)

# use simplify to see if sympy can figure out if the equation can be simplified:
display( sym.simplify(xab) ) # x^(a+b)

# substitue in specific numbers to the equation
n1 = 3.4
n2 = 7.3
# use subs to substitute a value anytime a symbol appears in the equation:
xa.subs(a,n1) * xb.subs(b,n2) # x^(10.7)

x**(n1+n2) # also x^10.7 - the same value from running the multiplication above showing this property of exponents
```

```python
# Now show that x^a/x^b = x^(a-b)

xa / xb # sympy automatically converts this to the equivalent multiplication op: x^a * x^-b (1/x^b)

sym.simplify(xa / xb) # x^(a-b)

n = 4
# Note: 0 ** 0 is 1 in python - this is just a convention and debated in mathematics.

(x**n) / (x**n) # 1 - anything divided by itself is equal to 1 (and this simplifies to x^0 because of the rule)
```

```python
# RULE: show (x^a)^b = x^(ab)

(xa)**b

# There is another simplification method in sympy for exponents: powsimp()
sym.powsimp((xa)**b) # note that this still does not simplify to x^(ab) and leaves the expression unaltered
# sympy simplifies only when the rule is generally true in the general case - see below constraints requirement on positive only values

n1 = 3
n2 = 7
display( (x**n1)**n2 ) # using specific numbers does simplify to the expected expression, x^21


n1 = 4.1
n2 = -.3
display( (x**n1)**n2 ) # using these numbers does NOT simplify as above

# now try with a negative base
y = -2
display( (y**n1)**n2 )
# in simplified form manually:
display( y**(n1*n2) ) # completely different number from above calculation

## ***** This shows that this rule does not always apply - it is true only when the two numbers are positive

# Create constraints so that we only use positive numbers by passing in positive argument to symbols()
y, c, d = sym.symbols('y,c,d',positive=True)
display( (y**c)**d ) # this will now show the correct simplification whereas above we couldn't get that without the constraint that makes the rule always hold
```

```python
##### RULES OF LOGS ######


# RULE: ln(a*b) = ln^a + ln^b

# set random number within a set: generates an array of 100 random numbers drawn from a uniform distribution over the interval [0, 1)
a = np.random.rand(100) # rand() generates nums between 0 and 1, so they will always be positive. randn() on the other hand could give you negative numbers, but you can't take the log of a negative number in the real domain
b = np.random.rand(100) # an alternative is to square all the numbers to make sure they are positive: ex: np.random.randn(100)**2

np.log(a*b) == np.log(a) + np.log(b)
# In python a lot of times you subtract the equation to expect 0 as the result to prove it is equivalent:
v = np.log(a*b) - ( np.log(a) + np.log(b) ) # expect to see result of 0 to show that the terms are equal. Note it's possible to get very small numbers due to computer rounding precision errors

# the 1e-16 at the top of the plot is saying that the scale is -1 to 2 * 1e-16 (very small rounding errors)
plt.plot(v, 's-') # all the results are based around 1e-16 showing they are all rounding errors and equivalent to 0
```

```python
# RULE: ln(a^b) = b * ln(a)

# as before set the equation to 0, and confirm that the answer is 0 with the plt
v = np.log(a**b) - b*np.log(a)

plt.plot(v, '-s') # the 1e-16 at the top of the plot is saying that the scale is -1 to 2 * 1e-16 (very small rounding errors)
```

```python
# RULE: ln(a / b) = ln(a) - ln(b)

v = np.log(a/b) - ( np.log(a) - np.log(b) ) # the expressions should balance to 0 to prove they are equivalent as before

# confirm the plot to show the results are all zero (or practically zero with rounding errors)
plt.plot(v, '-s')
```
