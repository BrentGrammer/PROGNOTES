```python
############# TRIG FUNCTIONS ##################

import numpy as np
import matplotlib.pyplot as plt

# define x axis of the grid/plot that has the range of pi's with linearly spaced numbers from -pi to 2pi
num_steps = 1234 # one thousand two hundred thirty four steps (numbers that will be produced below)
xx = np.linspace(-np.pi, 2*np.pi, num_steps) # -pi to 2pi radians on the x axis in num_steps number of ticks above, i.e. -3.14 radians (-180 degrees) to 6.28 (3+ times 360 degrees - multiple rotations) on the x-axis

########### Looking at the effects of squaring incorporated into the sin function ############

# initialize a list of 3 elements (0s)
f = [0] * 3

# replace each element to include a numpy array of the vector/output of a function
  # add a label as a second element in a tuple

f[0] = np.sin(xx) , 'sin(x)' # first function results in output of num_steps numbers with each as the input
# output of second function goes in next element of the list
f[1] = np.sin(xx)**2 , 'sin(x)**2' # note in math you write the ^2 square after sin, but in python it has to go on the end
# output of 3rd function - square the angle:
f[2] = np.sin(xx**2) , 'sin(x**2)'

# plot each of the function outputs:

# setting the above up as a tuple with values,label allows us to easily access in the loop to plot
# y is the output of the function
for y, label in f:
    plt.plot(xx,y,label=label)

plt.legend()
plt.xlabel('Angle (theta) in Radians: 1 =~ 57 deg')
plt.ylabel('y = f(x)')
plt.show()

# x-axis: each length of 2 represents 2pi (radians or 360 degrees, i.e. one full rotation around the unit circle)

# The blue line is the regular/base sin wave
# The orange line is taking the sin wave and squaring the output 
    # - note how you do not get any negative values (0 to 1 on y axis) due to the square, and the frequency is doubled
    # visually this represents going back and forth on one half of the unit circle (instead of around the whole circle as with regular sin above)
    # See video at https://www.udemy.com/course/pycalc1_x/learn/lecture/33947214 in timestamp 8:20 for visual explanation
# The green line is when the angle is squared that's passed into sin. Note how the frequency increases as theta increases.
  # Because the angle is squared, sin (the projection onto the x plane as coordinate moves) changes faster due to the larger angle differences at each step!
  # squaring something inside these trig functions will result in more interesting less symetrical results
```

```python
#### More trig function analysis - composing trig functions (composite functions) and more #####

# define x axis of the grid/plot that has the range of pi's with linearly spaced numbers from -pi to 2pi
num_steps = 1234 # one thousand two hundred thirty four steps (numbers that will be produced below)
xx = np.linspace(-np.pi, 2*np.pi, num_steps) # -pi to 2pi radians on the x axis in num_steps number of ticks above, i.e. -3.14 radians (-180 degrees) to 6.28 (3+ times 360 degrees - multiple rotations) on the x-axis

# initialize a list of 3 elements (0s)
f = [0] * 3

f[0] = np.sin(np.cos(xx)) , 'sin(cos(x))' # composing sine with cosine
f[1] = np.cos(np.sin(xx)) , 'cos(sin(x))'
f[2] = np.cos(xx) , 'cos(x)'

# plot each of the function outputs:

for y, label in f:
    plt.plot(xx,y,label=label)

plt.legend()
plt.xlabel('Angle (theta) in Radians: 1 =~ 57 deg')
plt.ylabel('y = f(x)')
plt.show()

#### Results ####

# note that the x-axis ranges from -pi to 2pi, but the cosine of x (passed into sine, for ex.) ranges between -1 and +1
  # Cosine is a unitless ratio, but when you pass into sin, it becomes Radians! so the y-axis is no longer unitless, but represents Radians (as if they were angles) - see timestamp 12:30 in video lecture
  # for example, you can draw the unit circle on a piece of paper and see what happens when you take the sine of angles between -1 and +1 radians (on the y-axis) vs. the cosine of angles between -1 and +1 radians
```

```python
####### TANGENT Trig function #######

# fig to allow multiple plots to show
rows = 2
cols = 1
fig, (ax1, ax2) = plt.subplots(rows,cols, figsize=(10, 10))



# resolution is 10 to the power of 2,3,4 (100,1000,10000) - i.e. how many points to plot
for i in range(2,5):
    # go from 0 to 2 in the different magnitude of steps, 100 1000 10000 points to plot in each loop
    xx = np.linspace(0,2*np.pi,10**i)

    # plot the tangent for each group of points
    ax1.plot(xx,np.tan(xx),'.-',label=f'{10**i} points')
    # make a second plot so you can zoom in on x axis below
    ax2.plot(xx,np.tan(xx),'.-',label=f'{10**i} points')


ax1.set_title('Zoomed out')
ax1.set_xlabel('Angle (radians)')
ax1.set_ylabel('tan(x)')
ax1.legend()

ax2.set_xlabel('Angle (radians)')
ax2.set_ylabel('tan(x)')
ax2.legend()

# zoom in on the x-axis so the grid space is 1.55 to 1.6 radians (somewhere around pi / 2 radians)
ax2.set_title('Zoomed in on x-axis')
ax2.set_xlim([1.55,1.6])

### ***NOTE: ***** IMPORTANT! ***** ###
#   some of the features of a continuous function as seen in a lower resolution graph plot are missing as we zoom in (there are less points plotted that don't reveal behavior that a higher number of points plotted would!!)
  # Note how the blue line (100 points vs. 1000 or 10000) does not capture the behavior and shapes of the plots with higher resolution points!
  # The yellow line captures a movement downwards, but is completely missing perturbations up and down seen in the higher resolution green line.

# Another side NOTE: As you go up to higher resolutions (more points plotted in the for loop, i.e. 10^8 etc.), the y-axis range will increase and approach infinity in either direction
```
