```python
# Built in graphics python lib tk
import tkinter as tk

root = tk.Tk()
win = tk.Canvas(root, width=1000, height=1000, bg="white")
win.pack()

# Start a line at point 0,0 and end at 100,100 (from x,y to x,y)
win.create_line(0,0,100,100,fill='blue')
win.create_polygon(100,100,100,200,200,100,200,200,fill='red')
# give the upper left corner and lower right corner of the oval (as if it were in a square)
win.create_oval(100,100,200,200,fill='green')

root.mainloop()
```

```python
# The Cantor Set

import tkinter as tk

# Uses recursion to create the set
def cantor_set(x,y,line):
    ONE_PIXEL = 1
    if line > ONE_PIXEL:
        win.create_line(x,y,x+line,y) # draw flat horizontal line
        y = y + 50 # move the next lines down on the screen
        cantor_set(x,y,line/3) # draw the left line 1/3 of the length of the previous line
        cantor_set(x+2/3*line,y,line/3) # right line starts 2/3rds len of the line over to the right


root = tk.Tk()
w,h = 1500,500
win = tk.Canvas(root, width=w, height=h)
win.pack()

margin_pixels = 20
# Start the line 10px to the right and 10px down from top of screen
cantor_set(10,10,w - margin_pixels) # 10px margin on either side of the starting line

root.mainloop()
```
