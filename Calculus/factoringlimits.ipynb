```python
########## SOLVING LIMITS WHERE THE PLUGIN METHOD DOES NOT WORK (RESULTS IN DIVIDING BY ZERO) #########

import numpy as np 
import sympy as sym 
import matplotlib.pyplot as plt 

from sympy.abc import x 

a,b = np.random.randint(-4,5,2) # 2 random numbers from -4 to 4 (exclusive upper bound of 5)
c = np.random.choice((-1,1)) # randomly choose -1 or 1 for c


###########################
#      USING FACTORING    #
###########################

# Factorized equation
expr = ( c*(x-a)*(x-b) / ( (x-a)*(x+b)) )
display('Factorized before expansion:')
display(expr)

# to make the exercise more involved we can expand the numerator and denominator
# We want to compute the limit as x goes to a of the following expression. (you can rerun the cell to get an infinite number of practice problem expressions generated)
expr = sym.expand( c*(x-a)*(x-b) ) / sym.expand( (x-a)*(x+b)) 
print('\nExpanded:')
display(expr)

```

```python
# Plot the expression

sym.plot(expr, (x, -10,10), ylim=[-10,10]) # set x domain to be -10 to +10 and the y limit as well for better viewing zoomed in
```

```python
# The limit from the left
left_limit = sym.limit(expr,x,a,dir='-')
right_limit = sym.limit(expr,x,a,dir='+')
limit = sym.limit(expr,x,a,dir='+-')
print(f'The limit as x->{a} from the left is: {left_limit}')
print(f'The limit as x->{a} from the right is: {right_limit}')
print(f'The limit as x->{a} (two-sided): {limit}')
print(f'Function value at x->{a}: {expr.subs(x,a)}') # NaN does not exist
```

```python
########## Another Exercise to generate random problems for calculating Limits ##############

##########################################################################################
# Using square roots and conjugates because the plugin method does not work for this eq. #
##########################################################################################

a = np.random.randint(-4,5) # a is limited to the set of numbers between -4 and 4 (5 excl.)
b = np.random.randint(a,a+5) # limited to a and up to a + 4 (5 excl.)
c = -sym.sqrt(b-a) # use sympi to get the theoretical actual square root repr. and not a numpy approximation

# An expression that we can't use the plugin method to compute the limit as x goes to negative a - will be 0/0
expr = ( sym.sqrt(x+b) + c) / (x+a)
display('Expression:')
display(expr)

print('Computing the Limit as x goes to negative a')

sym.plot(expr,(x,-10,10))
```

```python

# Show calculation of the limit as x goes to -a
left_limit = sym.limit(expr,x,-a,dir='-')
right_limit = sym.limit(expr,x,-a,dir='+')
limit = sym.limit(expr,x,-a,dir='+-')
print(f'The limit as x->{-a} from the left is: {left_limit}')
print(f'The limit as x->{-a} from the right is: {right_limit}')
print(f'The limit as x->{-a} (two-sided): {limit}')
print(f'Function value at x->{a}: {expr.subs(x,-a)}') # NaN does not exist
```

```python
############## EXPRESSIONMS WITH ABSOLUTE VALUES THAT RESULT IN 0 / 0 ######################

a, b = np.random.randint(-4,5,2)

expr = sym.Abs(x-a) + b # general form of abs value expression (has a limit)

display(f'Computing the limit as x goes to {a} of:')
display(expr)

sym.plot(expr,(x,10,-10))
# Note how functions with an absolute value operation like this will have a hard edge and a v-shape
```

```python
left_limit = sym.limit(expr,x,a,dir='-')
right_limit = sym.limit(expr,x,a,dir='+')
limit = sym.limit(expr,x,a,dir='+-')
print(f'The limit as x->{a} from the left is: {left_limit}')
print(f'The limit as x->{a} from the right is: {right_limit}')
print(f'The limit as x->{a} (two-sided): {limit}')
print(f'Function value at x->{a}: {expr.subs(x,a)}') # Note that we do have a limit that exists in this case
```
