```python
# Quadratic Equations plotting

import numpy as np
import matplotlib.pyplot as plt

steps = 3
start = -1
end = 3

xx = np.linspace(start,end,steps)

# negating the function: -ax^2 + bx + c
def fx(x, constants = {'a':1,'b':2,'c':3}): 
    return -constants['a']*(x**2) + constants['b']*x + constants['c']

# Changing 'a' affects the width and direction of the parabola
# Changing 'b' shifts the parabola horizontally
# Changing 'c' shifts the parabola vertically
constants = {
    'a': 1,
    'b': 1,
    'c': 1
}

plt.plot(xx, fx(xx, constants))

plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

xx = np.linspace(-1, 1, 5)  # Increase number of points for smoother curves

def fx1(x): return -x**2  # Base shape (inverted parabola)
# Next step is to add randomness to the transformations
def add_bump(y, amplitude=0.2, center=0, width=0.5):
    return y + amplitude * np.exp(-((xx-center)/width)**2)

def add_triangle(y, height=0.3, center=0, width=0.5):
    triangle = np.maximum(0, height - np.abs(xx - center) / width * height)
    return y + triangle

# List to store all transformations
transformations = [fx1(xx)]

# Apply transformations
transformations.append(add_bump(transformations[-1]))
transformations.append(add_triangle(transformations[-1], center=0.5))
transformations.append(add_bump(transformations[-1], amplitude=0.1, center=-0.5, width=0.3))


plt.figure(figsize=(12, 8))
for i, y in enumerate(transformations):
    plt.plot(xx, y, label=f'Transformation {i}')

plt.legend()
plt.title('Iterative Shape Transformations')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
